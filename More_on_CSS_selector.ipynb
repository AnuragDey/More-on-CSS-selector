```python
a{
    color: red;
}

header>nav>h2 {
    color: white;
    background-color: red;
}

nav>p:first-of-type {
    color: red;
}

a[href] {
    color: red;
    text-decoration: none;
}

a.active {
    color: blue;
    text-decoration: underline;
}

a[target="_self"] {
    background-color: aquamarine;
}

p:last-of-type::after {
    content:"...";
}

ul li::marker {
  color: purple;
}

p:nth-of-type(2)::first-letter {
  color: gray;
  font-size: 30px;
}
```
